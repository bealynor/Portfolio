# LendingClub Dataset
**Dataset Description：**
The dataset contains the homeloan records released from Lending Club. The problem lenders care is that whether a borrower can repay homeloan and interest on time. In addition, the homeloan interest rate is highly related with the loan credit of borrowers. A borrower with a higher credit can easily get homeloan with a lower rate. The task of the project is to predict whehter a borrower can replay homeloan and interest on time based on a number of features of the borrower. This is a typical classification problem and this notebook demonstrate how to use Lending Club dataset to decide whether a homeloan case should be approved based a borrower' features. 

**Dataset Detail Information*
The dataset contains 9,578 recrds in total. Every record has 13 features and 1 label which are described as below. 
* credit.policy: This is the label. Its value is 1 if the customer meets the credit underwriting criteria of LendingClub.com, and 0 otherwise.

* purpose: The purpose of the loan (takes values "creditcard", "debtconsolidation", "educational", "majorpurchase", "smallbusiness", “home_improvement” and "all_other").

* int.rate: The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers judged by LendingClub.com to be more risky are assigned higher interest rates.

* installment: The monthly installments owed by the borrower if the loan is funded.

* log.annual.inc: The natural log of the self-reported annual income of the borrower.

* dti: The debt-to-income ratio of the borrower (amount of debt divided by annual income).

* fico: The FICO credit score of the borrower.Common FICO scores range from 300 to 850,with higher scores indicating better credit.

* days.with.cr.line: The number of days the borrower has had a credit line.

* revol.bal: The borrower's revolving balance (amount unpaid at the end of the credit card billing cycle).

* revol.util: The borrower's revolving line utilization rate (the amount of the credit line used relative to total credit available).

* inq.last.6mths: The borrower's number of inquiries by creditors in the last 6 months.

* delinq.2yrs: The number of times the borrower had been 30+ days past due on a payment in the past 2 years.

* pub.rec: The borrower's number of derogatory public records (bankruptcy filings, tax liens, or judgments

* not.fully.paid：Whether the borrower will be fully paid or not.

**Learning taks:**
1. Data should be preprocessed and cleaned. 
2. Feature selection should be conducted to remove irrelevant features.
3. Train a logistic regression model to predict "credit.policy" by using the other 13 features.
4. The logistic regression model should be evaluated with cross validation by using 5-10 folds.

## Data Preprocess

### Process of Object data type

The logistic regression model cannot well process the object data type. We convert this data type with OneHotEncoder such that this feature can be handled by the logistic regression model. 

```python
#Display unique variables under 'purpose' feature/column
clean_data2['purpose'].unique()
```

```python
#convert categorical variables tagged as objects from the column "purpose" to dummy codes
dummy_purpose = pd.get_dummies(clean_data2['purpose'])
dummy_purpose
```

```python
#combine dummy_purpose dataframe with clean_data2 dataframe
clean_data3 = pd.concat((clean_data2.iloc[:,0],dummy_purpose, clean_data2.iloc[:,2:]), axis=1)

#show the new dataframe
clean_data3
```

### Dataset classification
We classify all data records into training set (80%), validation set (10%)  and test set (10%) so that we can determine hyper-parameters with k-cross validation. 

Randomly select 90% as the training + validation sets. The rest 10% will be used as the test set. 

```python
#count of credit policy (yes/no)
from sklearn.model_selection import train_test_split
print(clean_data3['credit.policy'].value_counts())
```

We complete dataset classification as below. 

```python
#classify data variables into training and test set values
x_training_set = clean_data3.drop(columns=['credit.policy'])
y_target_set = clean_data3.copy()['credit.policy']
x_training_array = x_training_set.values
y_target_array = y_target_set.values
```

```python
# split the dataset into training and test sets (80% for training and 20% for testing and validation)
X_train, X_test, y_train, y_test = train_test_split(x_training_array, y_target_array, test_size=0.2, random_state=42)

# split again 10% for testing and 10% for validation)
X_val, X_test,y_val, y_test,  = train_test_split(X_test, y_test, test_size=0.5, random_state=42)

# print the shapes of the resulting sets
print("Training set shape:", X_train.shape, y_train.shape)
print("Validation set shape:", X_val.shape, y_val.shape)
print("Test set shape:", X_test.shape,y_test.shape)
```

### Data normalisation

## Task 4: Recall that we have observed large value discrepancies between these features. It is necessary to normalise these features before we use them to train our models.  Here, we emply standardization method to normalise our dataset as below. 

```python
# standardize the dataset using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# print the shapes of the resulting standard datasets
print("Training set shape:", X_train_scaled.shape, y_train.shape)
print("Validation set shape:", X_val_scaled.shape, y_val.shape)
print("Test set shape:", X_test_scaled.shape, y_test.shape)
```

## Model Evaluation
In this stage, we are going to train a logistic regression model. Cross validation will be used to determine hyper-parameters and evaluate model performance.


### Logistic Regression Model

## Task 5: Train a Logistic Regression Model with training dataset.

```python
# Train the model on the training data
model_log = LogisticRegression()
model_log.fit(X_train_scaled, y_train)

#Predict the labels for the test data
y_pred = model_log.predict(X_test_scaled)
```

```python
#Create variable for the Number of splits
kf = KFold(n_splits=10)

# Create variable for all the performance metrics to be used in cross-validation
scoring = ['accuracy', 'recall', 'f1', 'precision']

# Perform cross-validation and get the performance metrics for each fold
scores = cross_validate(model_log, X_train_scaled, y_train, cv=kf, scoring=scoring)

#Compute the accuracy performance
accuracy = scores ['test_accuracy']
recall = scores ['test_recall']
precision = scores ['test_precision']
f1 = scores ['test_f1']

#Compute the mean scores for each performance metric
mean_accuracy = np.mean(accuracy)
mean_precision = np.mean(precision)
mean_recall = np.mean(recall)
mean_f1 = np.mean(f1)

#Print the mean of each performance metrics and its standard deviation 
print("Accuracy: %0.3f (+/- %0.2f)" % (mean_accuracy, accuracy.std() * 2))
print("Recall: %0.3f (+/- %0.2f)" % (mean_recall,recall.std() * 2))
print("F1 Score: %0.3f (+/- %0.2f)" % (mean_f1,f1.std() * 2))
print("Precision: %0.3f (+/- %0.2f)" % (mean_precision,precision.std() * 2))
```


## Task 6: To better understand our result, we visualize the performance evluation by comparing the model accuracy of the Logistic Regression model on the training dataset and each validation dataset.

```python
# Define the number of folds
number_folds = 10

# Create a list of x-axis labels for each fold
fold_labels = ["Fold {}".format(i) for i in range(1, number_folds+1)]

# Create a list of y-axis values for each metric and each fold
accuracy_values = accuracy
precision_values = precision
recall_values = recall
f1_values = f1

# Create the bar graph
bar_width = 0.2
opacity = 0.8

plt.bar(np.arange(len(fold_labels)), accuracy_values, bar_width, alpha=opacity, color='sandybrown', label='Accuracy')
plt.bar(np.arange(len(fold_labels)) + bar_width, precision_values, bar_width, alpha=opacity, color='royalblue', label='Precision')
plt.bar(np.arange(len(fold_labels)) + 2*bar_width, recall_values, bar_width, alpha=opacity, color='darkseagreen', label='Recall')
plt.bar(np.arange(len(fold_labels)) + 3*bar_width, f1_values, bar_width, alpha=opacity, color='orchid', label='F1 Score')


# Display the mean scores as horizontal lines
plt.axhline(mean_accuracy, color='b', linestyle='--')
plt.axhline(mean_precision, color='g', linestyle='--')
plt.axhline(mean_recall, color='r', linestyle='--')
plt.axhline(mean_f1, color='y', linestyle='--')

# Add labels and title for x & y axis
plt.xlabel("Folds")
plt.ylabel("Score")
plt.title("Model Performance Metrics (10-Fold Cross-Validation)")
plt.xticks(np.arange(len(fold_labels)) + 2*bar_width, fold_labels)
plt.legend()
plt
```

```python
#Use the score method to evaluate the performance of the logistic regression model on the training data
train_score = model_log.score(X_train_scaled, y_train)
print("Training Accuracy:", train_score)
```

```python
#Use the score method to evaluate the performance of the logistic regression model on the validation data
validation_score = model_log.score(X_val_scaled, y_val)
print("Validation Accuracy:", validation_score)
```

```python
#create a bargraph to visualize accuracy of training and validation datasets

def plot_accuracy (model, x_train, y_train, x_val, y_val):
    train_acc = accuracy_score (y_train, model.predict(X_train_scaled))
    val_acc = accuracy_score (y_val, model.predict(X_val_scaled))
    colors = ['pink', 'mediumturquoise']
    width = 0.1
    plt.bar(['Training Accuracy', 'Validation Accuracy'], [train_acc, val_acc], color=colors, width = .5, align = "center")
    plt.title ('Logistic Regression Model Accuracy')
    plt.show()
    
    return train_acc, val_acc

#plot the training and validation sets to the bargraph
train_acc, val_acc = plot_accuracy (model_log, X_train_scaled, y_train, X_val_scaled, y_val)
print(f"Training Accuracy: {train_acc}")
print(f"Validation Accuracy: {val_acc}")
    
```
